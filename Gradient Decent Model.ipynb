```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Generate some random data for demonstration
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Plot the data
plt.scatter(X, y)
plt.xlabel('X')
plt.ylabel('y')
plt.title('Data points')
plt.show()

# Define the learning rate and number of iterations
learning_rate = 0.1
num_iterations = 100

# Initialize the parameters (theta_0 and theta_1)
theta_0 = np.random.randn()
theta_1 = np.random.randn()

# Perform gradient descent
for iteration in range(num_iterations):
    # Compute the predicted values
    y_pred = theta_0 + theta_1 * X
    
    # Compute the gradients
    gradient_0 = (1/len(X)) * np.sum(y_pred - y)
    gradient_1 = (1/len(X)) * np.sum((y_pred - y) * X)
    
    # Update the parameters
    theta_0 -= learning_rate * gradient_0
    theta_1 -= learning_rate * gradient_1

# Plot the data and the linear regression line
plt.scatter(X, y)
plt.plot(X, theta_0 + theta_1 * X, color='red')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression with Gradient Descent')
plt.show()

# Print the learned parameters
print(f'Learned theta_0: {theta_0}')
print(f'Learned theta_1: {theta_1}')
```
